```python
import pandas as pd

df = pd.read_csv("C:\export\data.csv",delimiter=";")


dataset = df.ix[:,[1,3,4]]

dataset = dataset.values

print(dataset)

print("#################")

dflabel = pd.read_csv("C:\export\label.csv",delimiter=";",header=None)

label = dflabel.ix[:,[1,2,3]]

label = label.values

print(label)




```

```python
#import pandas

import pandas as pd
import pickle

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

from sklearn import preprocessing

# split X and y into training and testing sets
from sklearn.cross_validation import train_test_split


X_train,X_test,y_train,y_test=train_test_split(dataset,label,test_size=0.20,random_state=0)




# normalize the data attributes


# Create linear regression object
regr = linear_model.LinearRegression()



# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)


filename = 'linear_reg.model'

pickle.dump(regr, open(filename, 'wb'))

loaded_model = pickle.load(open('linear_reg.model', 'rb'))


print('hero')
print(regr.predict([[250.083984375, 0.00017595492516951105, 1096.5136571223536]]))


y_pred1 = loaded_model.predict(X_test)

print(y_pred1)

# The coefficients
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test,y_pred))


import matplotlib.pyplot as plt
plt.figure(1)
plt.title('Plombe: Predictions vs Valeur reelle')
plt.ylabel('Plombe')
plt.xlabel('Image')


plt.plot(np.arange(1,8,1),y_pred[:,0], 'ro-')
plt.plot(np.arange(1,8,1),y_test[:,0], 'bo-')
plt.legend(['Predicted','Real'], loc='upper right')
plt.show()


import matplotlib.pyplot as plt
plt.figure(1)
plt.title('Cuivre: Predictions vs Valeur reelle')


plt.ylabel('Cuivre')
plt.xlabel('Image')


plt.plot(np.arange(1,8,1),y_pred[:,1], 'ro-')
plt.plot(np.arange(1,8,1),y_test[:,1], 'bo-')
plt.legend(['Predicted','Real'], loc='upper right')
plt.show()



import matplotlib.pyplot as plt
plt.figure(1)
plt.title('Zn: Predictions vs Valeur reelle')
plt.ylabel('Zn')
plt.xlabel('image')


plt.plot(np.arange(1,8,1),y_pred[:,2], 'ro-')
plt.plot(np.arange(1,8,1),y_test[:,2], 'bo-')
plt.legend(['Predicted','Real'], loc='upper right')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor


# Create a random dataset

max_depth = 30

regr_multirf = MultiOutputRegressor(RandomForestRegressor(n_estimators=100,
                                                          max_depth=max_depth,
                                                          random_state=0))
regr_multirf.fit(X_train, y_train)

print("Score is : "+str(regr_multirf.score(X_train, y_train)))

regr_rf = RandomForestRegressor(n_estimators=100, max_depth=max_depth,
                                random_state=2)
regr_rf.fit(X_train, y_train)

# Predict on new data
y_multirf = regr_multirf.predict(X_test)
y_rf = regr_rf.predict(X_test)

# Plot the results
plt.figure()
s = 50
a = 0.4
plt.scatter(y_test[:, 0], y_test[:, 1], edgecolor='k',
            c="navy", s=s, marker="s", alpha=a, label="Data")
plt.scatter(y_multirf[:, 0], y_multirf[:, 1], edgecolor='k',
            c="cornflowerblue", s=s, alpha=a,
            label="Multi RF score=%.2f" % regr_multirf.score(X_test, y_test))
plt.scatter(y_rf[:, 0], y_rf[:, 1], edgecolor='k',
            c="c", s=s, marker="^", alpha=a,
            label="RF score=%.2f" % regr_rf.score(X_test, y_test))
plt.xlim([-100, 100])
plt.ylim([-100, 100])
plt.xlabel("target 1")
plt.ylabel("target 2")
plt.title("Comparing random forests and the multi-output meta estimator")
plt.legend()
plt.show()
```
